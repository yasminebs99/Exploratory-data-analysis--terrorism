```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
terrorism_data = pd.read_csv("Global Terrorism - START data/globalterrorismdb_0718dist.csv", encoding='ISO-8859-1', low_memory=False)

print('*******************************************************')
print('Display basic information about the dataset')
print(terrorism_data.info())

print('*******************************************************')
print('Display the first few rows of the dataset')
print(terrorism_data.head())

# Exploratory Data Analysis
print('*******************************************************')
print('Summary statistics')
print(terrorism_data.describe())

print('*******************************************************')
print('Check for missing values')
print(terrorism_data.isnull().sum())
```

```python
# Visualize distribution of terrorist attacks over years
plt.figure(figsize=(12, 6))
sns.histplot(data=terrorism_data, x='iyear', bins=range(1970, 2026), kde=True, color='skyblue')
plt.title('Distribution of Terrorist Attacks Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Visualize distribution of terrorist attacks by region
plt.figure(figsize=(12, 6))
sns.countplot(data=terrorism_data, x='region_txt', order=terrorism_data['region_txt'].value_counts().index, palette='viridis')
plt.xticks(rotation=90)
plt.title('Distribution of Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Find hot zone of terrorism (region with highest number of attacks)
hot_zone = terrorism_data['region_txt'].value_counts().idxmax()
print(f"The hot zone of terrorism is: {hot_zone}")
```
